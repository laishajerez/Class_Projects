```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
# 'School_A' and 'School_B' contain the name of the schools I will compare
school_a = 'Academy of Finance and Enterprise'
school_b = 'Bard High School Early College'
```

```python
# Filter (rows) data for each school
data_school_a = df[df['School Name'] == school_a]
data_school_b = df[df['School Name'] == school_b]
```

```python
# print the data for school a 'Academy of Finance and Enterprise'
print("Data for School A:")
print(data_school_a)
```

```python
# print the data for school b 'Bard High School Early College'
print("\nData for School B:")
print(data_school_b)
```

```python
# Shows 'Regents Exam' for Academy of Finance and Enterprise (school A) by the 'Years' and 'Mean Score' and Print it out
print(f"{school_a} Regents Exam:")
print(data_school_a[['Year', 'Regents Exam', 'Mean Score']])
```

```python
# Filter the data from the above code, showing only 'Common Core Algebra' Regent Exam by the 'Years' and 'Mean Score' and to print it out
exam_name = 'Common Core Algebra'
filtered_data_school_a = data_school_a[data_school_a['Regents Exam'] == exam_name]

# Show the filtered data that was asked
print(f"{school_a} {exam_name}:")
print(filtered_data_school_a[['Year', 'Regents Exam', 'Mean Score']])
```

```python
# Show the Mean of the Mean Scores for each 'Year'
filtered_data_school_a['Mean Score'] = pd.to_numeric(filtered_data_school_a['Mean Score'], errors='coerce')  # Convert 'Mean Score' to numeric

combined_data_school_a = filtered_data_school_a.groupby('Year')['Mean Score'].mean().reset_index()

# Display the combined data
print(f"{school_a} Common Core Algebra (Combined):")
print(combined_data_school_a)
```

```python
import matplotlib.pyplot as plt
# Show a illustration (graph) of the Mean Score of each year
plt.figure(figsize=(10, 6))
plt.plot(combined_data_school_a['Year'], combined_data_school_a['Mean Score'], marker='o', linestyle='-')

plt.title(f"{school_a} Common Core Algebra Mean Scores Over Years")
plt.xlabel('Year')
plt.ylabel('Mean Score')
plt.grid(True)
plt.show()
```

```python
# Shows 'Regents Exam' for Bard High School Early College (school B) by the 'Years' and 'Mean Score'
print(f"{school_b} Regents Exam:")
print(data_school_b[['Year', 'Regents Exam', 'Mean Score']])
```

```python
# Filter the data from the above code, showing only 'Common Core Algebra' Regent Exam by the 'Years' and 'Mean Score'
exam_name = 'Common Core Algebra'
filtered_data_school_b = data_school_b[data_school_b['Regents Exam'] == exam_name]

# Show the filtered data that was asked
print(f"{school_b} {exam_name}:")
print(filtered_data_school_b[['Year', 'Regents Exam', 'Mean Score']])
```

```python
# Show the Mean of the Mean Scores for each 'Year'
filtered_data_school_b['Mean Score'] = pd.to_numeric(filtered_data_school_b['Mean Score'], errors='coerce')  # Convert 'Mean Score' to numeric

combined_data_school_b = filtered_data_school_b.groupby('Year')['Mean Score'].mean().reset_index()

# Display the combined data
print(f"{school_b} Common Core Algebra (Combined):")
print(combined_data_school_b)
```

```python
import matplotlib.pyplot as plt
# Show a illustration (graph) of the Mean Score of each year
plt.figure(figsize=(10, 6))
plt.plot(combined_data_school_b['Year'], combined_data_school_b['Mean Score'], marker='o', linestyle='-')

plt.title(f"{school_b} Common Core Algebra Mean Scores Over Years")
plt.xlabel('Year')
plt.ylabel('Mean Score')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine data for school A and school B
combined_data_schools = pd.concat([data_school_a.assign(School=school_a),
                                   data_school_b.assign(School=school_b)])

# Convert 'Mean Score' to numeric for both schools
combined_data_schools['Mean Score'] = pd.to_numeric(combined_data_schools['Mean Score'], errors='coerce')

# Plot the mean score
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='Mean Score', hue='School', data=combined_data_schools, palette='viridis')

plt.title("Common Core Algebra Mean Scores Between AOFE and Bard High School Over Years")
plt.xlabel('Years')
plt.ylabel('Mean Score')
plt.legend(title='School')
plt.show()
```
